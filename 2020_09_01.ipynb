```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
Ome0 = 1
Ome02 = Ome0**2
deltaT = .01
Tmax = 100

t = np.arange(0, Tmax, deltaT)
N = len(t)

teta = np.zeros(N)
w = np.zeros(N)
Energia = np.zeros(N)

teta[0] = 20/np.pi/180
w[0] = 0
```

```python
for i in range(N-1):
    w[i+1] = -Ome02*np.sin(teta[i])*deltaT+w[i]
    teta[i+1] = teta[i]+w[i]*deltaT
```

```python
plt.figure(1)
plt.plot(t,teta,'k',t,w,'r',linewidth=2)
plt.xlabel('t(s)', fontsize=18)
plt.ylabel('teta em preto, omega em vermelho', fontsize=18)
```

```python
plt.figure(2)
plt.plot(teta,w,'r',linewidth=1)
plt.xlabel('theta', fontsize=18)
plt.ylabel('omega', fontsize=18)
```

```python
plt.figure(3)
Energia = w**2/2 + Ome02*(1-np.cos(teta))
plt.plot(t,Energia,'b',linewidth=2)
plt.xlabel('t(s)')
plt.ylabel('Energia(u.a.)')
```

```python
fig = plt.figure(4)
ax = fig.add_subplot(111, projection='3d')
ax.plot(teta,w,t)
```
